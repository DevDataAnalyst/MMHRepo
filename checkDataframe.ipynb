```python
import pandas as pd
```

```python
Df1 = pd.DataFrame({'name': ['Marc', 'Jake', 'Sam', 'Brad','John', 'Tony', 'Bob']})
Df2 = pd.DataFrame({'IDs': ['Jake', 'John', 'Marc', 'Tony', 'Bob']})
Df3 = pd.DataFrame({'ls': ['Sam']})
```

```python
Df1.assign(InDf2=Df1.name.isin(Df2.IDs).astype(int))
```

```python
s=Df1.merge(Df2,left_on='name',right_on='IDs',how='left')
s.IDs=s.IDs.notnull().astype(int)
s
```

```python
lv = Df3.merge(s,left_on='ls',right_on='name',how='right')
lv.ls=lv.ls.notnull().astype(int)
lv
```
